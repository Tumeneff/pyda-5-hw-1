```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from scipy.misc import derivative #для подсчета производной 

```

**Задание 1**

Посчитайте значение первой и второй производной функции:

f(x) = x^5 + 4sin(2x) + cos(3x+3)
        
В точке x = 1

```python
# задаем функцию
def func1(x):
    return x**5+4*np.sin(2*x) + np.cos(x*3+3)
```

```python
print('Производная 1 порядка: ', derivative(func1, 1.0, n=1))# считаем производную, указывая функцию, значение х (1.0) и порядок производной (n=1)
```

```python
print('Производная 2 порядка: ', derivative(func1, 1.0, n=2)) # считаем производную, указывая функцию, значение х (1.0) и порядок производной (n=2)
```

**Задание 2**

Постройте график первой и второй производной функции:

f(x) = sin^5(2x+1)

На отрезке [-5; 5].

Для начала просто решил посмотреть как выглядит график самой функции:

```python
x = np.arange(-5, 5, 0.1)
y = np.sin(2*x+1)**5
plt.plot(x,y)
# plt.plot(x,func1(x))
plt.show
```

Теперь зададим функцию и также создадим два списка, содержащие значения производных первого и второго порядка для заданного диапазона, а потом построим графики с использованиемзнчений из этих списков:

```python
def func_2(x):
    return np.sin(2*x+1)**5
x = np.arange(-5, 5, 0.1)
f = [] # список значений производной первого порядка
s = [] # список значений производной второго порядка

for i in x:   
    f.append(derivative(func_2, i, n=1))
    s.append(derivative(func_2, i, n=2))

plt.figure(figsize=(15, 10))
plt.plot(x,f, label = 'график производной 1-го порядка') 
plt.plot(x,s, label = 'график производной 2-го порядка')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show
```
