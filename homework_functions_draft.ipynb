```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
# Задание 1
def search_bn(docs): # функция, ищущая владельца по номеру док-та
    search_by_number = input("Введите номер:\n ")
    name_x = ''
    for el in docs:
        if search_by_number == el['number']:
            name_x = el['name']
            return True, name_x
    return False

def searchbynumout(docs): # функция, выводящая результат работы search_bn(docs)
    finded, nx = search_bn(docs)
    if finded:
        print('Владелец документа: ', nx)
    else:
        print('Документ не найден в базе')


# Задание 2
def search_bd(dirs): # функция, ищущая документ на полках
    search_by_doc_n = input("Введите номер:\n ")
    for dic_k, dic_v in dirs.items():
        for el in dic_v:
            if search_by_doc_n == el in dic_v:
                out_v= dic_k
                return True, out_v
    return False

def searchbydocout(dirs): # функция, выводящая владельца документа по результатам работы функции search_bd(dirs)
    finded, sf = search_bd(dirs)
    if finded:
        print('Документ хранится на полке: ', sf)
    else:
        print('Документ не найден в базе')

# Задание 3      
def data_output(dirs, docs): # функция выводит итоговый список
    for k, v in dirs.items():
        for el in docs:
            if el['number'] in v:
                el['shelve'] = k
                print('ФИО: ', el['name'] +'.' ' Тип документа:', el['type']+'.' ' №:', el['number']+'.' ' Номер полки:',el['shelve']+'.')                      

# Задание 4
def dirs_quantity(dirs): # функция, выводящая количество полок
    d_q = ', '.join(list(dirs.keys()))
    return d_q

def add_shelve_command(dirs): # функция, добавляющая полку
    add_shelve = input("Введите номер полки:\n ")
    if add_shelve == None:
        return None, None
    if add_shelve not in dirs.keys():
        dirs[add_shelve] = []
        dirs_n = dirs_quantity(dirs)
        return True, dirs_n
    return False, None

def addshelve_out(dirs): # функция, выводящая результат по итогам добавления полки
    finded, sf = add_shelve_command(dirs)
    if finded:
        print('Полка добавлена. Текущий перечень полок: ', sf)
    else:
        print('Такая полка уже есть, текущий перечень полок: ', dirs_quantity(dirs))

# Задание 5
    
def del_shelve_command(dirs): # функция, удаляющая полку
    del_shelve = input("Введите номер полки:\n ")
    for k, v in dirs.items():
        if del_shelve in k:
            if not v:
                del(dirs[k])
                dirs_n = dirs_quantity(dirs)
                return True, dirs_n              
            else:
                dirs_n = dirs_quantity(dirs)
                return None, dirs_n
    return False, None

def delshelve_out(dirs): # функция, выводящая результат по итогам удаления полки
    finded, sf = del_shelve_command(dirs)
    if finded:
        print('Полка удалена. Текущий перечень полок: ', sf)
    elif sf:
         print('На полке есть документы, удалите их перед удалением полки.')
    else:
        print('Такой полки не существует. Текущий перечень полок: ', dirs_quantity(dirs))


    
# def add_shelve_command(dirs): # громоздкая и недекомпозированная функция, чтобы добавить полку
#     add_shelve = input("Введите номер полки:\n ")
#     a = False
#     for k, v in dirs.items():
#         if add_shelve not in dirs.keys():
#             a = True
#             dirs[add_shelve] = []
#             print('Полка добавлена. Текущий перечень полок: ', ', '.join(list(dirs.keys())))
#             break        
#     if not a:
#         print('Такая полка уже есть, текущий перечень полок: ', ', '.join(list(dirs.keys())))
#     return dirs

# def del_shelve_command(dirs): # громоздкая и недекомпозированная функция, чтобы удалить полку
#     del_shelve = input("Введите номер полки:\n ")
#     a = False
#     for k, v in dirs.items():
#         if del_shelve in k:
#             a = True
#             if not v:
#                 del(dirs[k])
#                 print('Полка удалена. Текущий перечень полок: ', ', '.join(list(dirs.keys())))
#                 break        
#             else:
#                 print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ', ', '.join(list(dirs.keys())))
#     if not a:
#         print('Такой полки не существует. Текущий перечень полок: ', ', '.join(list(dirs.keys())))        
#     return dirs

# Задание 6
def add_doc(docs, dirs): # громоздкая и недекомпозированная функция, чтобы добавить документ
    doc_number = input("Введите номер документа:\n ")
    doc_type = input("Введите тип документа:\n ")
    a = False
    y = False
    for el in docs:
        if doc_number == el['number'] and doc_type == el['type']:
            a = True
            print('Cовпадение!!! В базе уже есть человек с такими данными:', el['name'])
            break
    if not a:
        doc_person = input("Введите ФИО человека:\n ")
        doc_shelve = input("Введите номер полки:\n ")
        for k, v in dirs.items():
            if doc_shelve in k:
                y = True
                docs.append({'type': doc_type, 'number': doc_number, 'name': doc_person})
                dirs[doc_shelve].append(doc_number)
                print('Документ добавлен. Текущий список документов:\n')
                for el in docs:
                    print('ФИО:', el['name'] +'.' ' Тип документа:', el['type']+'.' ' №:', el['number']+ '.')

        if not y:
            print('\033[91m\033[1m!!!Такой полки не существует!!!!\n Сначала создайте полку командой "ads" \033[0m\033[0m')
    return docs, dirs

# Задание 7
def deldocs(docs, dirs): # громоздкая и недекомпозированная функция, чтобы удалить документ
    del_doc = input("Введите номер документа:\n ")
    a = False
    for el in docs:
        for k, v in el.items():
            for s, n in dirs.items():
                if del_doc in v and del_doc in n:
                    a = True
                    docs.remove(el)
                  #  print(n)
                    n.remove(del_doc)
                    print('Документ удален. Текущий список документов:\n')
                    for el in docs:
                        print('ФИО:', el['name'] +'.' ' Тип документа:', el['type']+'.' ' №:', el['number']+ '.', ' Полка номер', s)

                    break        
    if not a:
        print('Такого документа нет')        
    return docs, dirs

def main():
    initial_input = input('введите команду: ')

    while initial_input != "x":
        if initial_input == "p":
            searchbynumout(documents)
            initial_input = input('введите команду: ')
        elif initial_input == "s":
            searchbydocout(directories)
            initial_input = input('введите команду: ')
        elif initial_input == "l":
            data_output(directories, documents)
            initial_input = input('введите команду: ')
        elif initial_input == "ads":
            addshelve_out(directories)
            initial_input = input('Введите команду: ')
        elif initial_input == "ds":
            delshelve_out(directories)
            initial_input = input('Введите команду: ')
        elif initial_input == "ad":
            add_doc(documents, directories)
            initial_input = input('Введите команду: ')
        elif initial_input == "d":
            deldocs(documents, directories)
            initial_input = input('Введите команду: ')
        else:
            print('Такой команды нет')
            initial_input = input('введите команду: ')
    if initial_input == "x":
            print('работа программы завершена')
main()
```
