```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

**Задание 1**

Посчитайте частные производные функций:

        1) f(x,y)=2x^2y^3 + 1/x + y^2x + 7,
       
        2) f(x,y)=x^2y - sin(xy) + cos(x^2) + 6y

```python
x = sym.Symbol('x')
y = sym.Symbol('y')
print('частная производная 1-ой функции по х:\n')
sym.diff(2*x**2*y**3 + 1/x+ y**2*x+7, x)
```

```python
print('частная производная 1-ой функции по y:\n')
sym.diff(2*x**2*y**3 + 1/x+ y**2*x+7, y)
```

```python
f = x**2*y - sym.sin(x*y)+sym.cos(x**2)+6*y, x
print('частная производная 2-ой функции по x:\n')
sym.diff(f, x)
```

```python
print('частная производная 2-ой функции по y:\n')
sym.diff(f, y)
```

**Задание 2**

Вычислите градиент функции:
        
        f(x,y)=2x^2y^3 + 1/x + y^2x + 7
        
        в точке (1,2)

```python
fun=2*x**2*y**3 + 1/x + y**2* + 7

#создаем переменную градиента
gradfun=[sym.diff(fun,var) for var in (x,y)]

# с помощью lymbdyfy выводим градиент 
gradient=sym.lambdify([x,y],gradfun)
gradient(1,2)
```

**Задание 3**

Постройте трехмерный график функции:
    
       f(x)=x_2^2-x_1^2 x_2 + 2 x_1 x_2

```python
x1 = np.arange(-5, 5, 0.25)
x2 = np.arange(-5, 5, 0.25)
z = x2**2-x1**2*x2+2*x1*x2
fig = plt.figure()
fig.set_size_inches(7, 7)
ax = fig.add_subplot(111, projection='3d')
ax.plot(x1, x2, z)
ax.set_title('график функции')
```
