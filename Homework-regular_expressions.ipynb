## Задание 1

Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером (1 буква, 3 цифры, 2 буквы, 2-3 цифры). Обратите внимание, что не все буквы кириллического алфавита используются в транспортных номерах.

Если номер валиден, то функция должна возвращать отдельно номер и регион.

Примеры работы программы:  

1. `car_id = 'АВ222С96'`

Результат: 
`Номер АВ222С валиден. Регион: 96`

2. `car_id = 'АБ22ВВ193'`

Результат: 
`Номер не валиден`

````python
import re
````

***Решение Задания 1***

````python
car_id = input('Введите номер авто: ') #принимаем номер авто
car_id = car_id.upper() #все символы в номере делаем верхним регистром
result = re.findall(r'[АВЕКМНОРСТУХ]\d{3}[АВЕКМНОРСТУХ]{2}\d{2,3}', car_id) #проверяем на валидность
no_region_num = re.findall(r'[АВЕКМНОРСТУХ]\d{3}[АВЕКМНОРСТУХ]{2}', car_id) #создаем переменную с номером без региона
region = re.findall(r'(?<=[АВЕКМНОРСТУХ]{2})\d{2,3}$', car_id)  #вычленям регион и загоняем в отдельную переменную
if result: #если номер валиден
    print('Номер {} валиден. Регион: {}'.format(no_region_num[0], region[0])) #выводим текст с номером и регионом
else:# если нет:
    print('Номер не валиден') # пишем, что номер кривой
````

## Задание 4

Напишите функцию, которая будет принимать на вход список email-адресов и выводить их распределение по доменным зонам.

Пример работы программы:  

`emails = ['test@gmail.com, xyz@test.in, test@ya.ru, xyz@mail.ru, xyz@ya.ru', xyz@gmail.com]`

Результат: 
```
gmail.com: 2
test.in: 1 
ya.ru: 2 
mail.ru: 1
```

***Решение Задания 4***

````python
emails = ['test@gmail.com, xyz@test.in, test@ya.ru, xyz@mail.ru, xyz@ya.ru, xyz@gmail.com']
def email_domains_counter(e_list):
    import re
    emails_s = str(e_list[0])
    match = re.findall(r'(\w+\.\w+)(?!.*\1)', emails_s)
    em_l = re.split(r'@|,', emails_s)
    match_d = {}.fromkeys(match, 0)
    for a in em_l:
        if a in match_d.keys():
            match_d[a] += 1
    for k, v in match_d.items():
        print('{}: {}'.format(k,v))

email_domains_counter(emails)   
````

Результат: 
```

```

***Решение Задания 5***

````python
some_text = 'Эталонной реализацией Python является интерпретатор CPython, поддерживающий большинство активно используемых платформ. Он распространяется под свободной лицензией Python Software Foundation License, позволяющей использовать его без ограничений в любых приложениях, включая проприетарные.'
def words_counter(text):
    import re
    text_s = str(text)
    alfa_v = re.findall(r'\b[АОУЫЭЁЮИЕЯаоуыэеёюияAEIOUYaeiouy]', text_s)
    alfa_cons = re.findall(r'\b[НМЛРЙБВГДЖЗПФКТШСХЧЦЩнмлрйбвгджзпкфштсхцчщBCDFGHJKLMNPQRSTVWXZbcdfghjklmnpqrstvwxz]', text_s)
    words = re.split(r'\s', text_s)
    words = re.split(r'\s', text_s)
    vowel = 0
    consonant = 0
    overall = len(words)
    for i in words:
        if i[0] in alfa_v:
            vowel +=1
        if i[0] in alfa_cons:
            consonant += 1
    print('Общее количество слов в тексте: ', overall)
    print('Слов на гласные буквы: ', vowel)
    print('Слов на согласные буквы: ',  consonant)
    return

words_counter(some_text)
````

***Решение Задания 6***

````python
def checkMnumber():
    mobile = input('Введите номер: ')
    import re
    #провера номера на валидность. Обязательно присутствие +7,8 или 7 в начале, после блок из трех цифр с возможными тире
    #, а дальше блок из 7 цифр с возможными тире
    mobile_check = re.findall(r'((8|\+7|7)[\- ]?)(\(?\d{3}\)?[\- ]?)([\d\- ]{7,9})', mobile) 
    a = '' #пустая переменная, в которую вы итоге будет записываться номер в правильном формате и которая будет ввозвращаться
    if mobile_check: #если прошли проверку на валидность
        number = re.sub(r'[\(\)\-\+\s]', r'', mobile) #удаляем все символы кроме цифр
        #записываем в переменную "a" номер в том виде, который нам требуется, используя поиндексно значения из списка number
        a = '+7-{}-{}-{}-{}'.format(number[1:4], number[4:7], number[7:9], number[9:]) 
        return a #возвращаем а
    return 'Номер не валиден' #возвращаем 'Номер не валиден', если проверка на валидность не прошла
checkMnumber()
````
