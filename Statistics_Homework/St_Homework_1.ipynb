Однако, самое веселое начинается, когдамы начинаем тестировать категории 2 и 1 (у вздутия это "отсутствует" и "легкое", а у перистальтики это "сниженная" и "нормальная" соответственно).

При любых значениях вздутия (говорим про 1, 2) перистальтика показывает 3 категорию: 

```python
print('При вздутии живота категории 1 медиана по перистальтики: ', horses_data[(horses_data['Вздутие живота'] == 1)]['Перистальтика'].median())
print('При вздутии живота категории 2 медиана по перистальтики: ', horses_data[(horses_data['Вздутие живота'] == 2)]['Перистальтика'].median())
print('При вздутии живота категории 3 медиана по перистальтики: ', horses_data[(horses_data['Вздутие живота'] == 3)]['Перистальтика'].median())
print('При вздутии живота категории 4 медиана по перистальтики: ', horses_data[(horses_data['Вздутие живота'] == 4)]['Перистальтика'].median())
```

***Однако, все меняется, когда мы делаем проверку вздутия по перистальтике***

```python
print('При перистальтике категории 1 медиана по вздутию: ', horses_data[(horses_data['Перистальтика'] == 1)]['Вздутие живота'].median())
print('При перистальтике категории 2 медиана по вздутию: ', horses_data[(horses_data['Перистальтика'] == 2)]['Вздутие живота'].median())
print('При перистальтике категории 3 медиана по вздутию: ', horses_data[(horses_data['Перистальтика'] == 3)]['Вздутие живота'].median())
print('При перистальтике категории 4 медиана по вздутию: ', horses_data[(horses_data['Перистальтика'] == 4)]['Вздутие живота'].median())
```

**Что же делать, как же быть?**

Исходя из логики, для малых значений (1, 2) правильнее будет взять за основу медианы вздутия по Перистальтике, а не наоборот.

Во-первых, ясно сказано, что при малых значниях вздутия перистальтика с высокой долей вероятности будет более выражена, что явно не попадает в категорию 3.

Во-вторых, данных по перистальтике несколько больше, чем по вздутию (255 против 243), что дает нам условно большую уверенностьв правильности за счет несколько большей выборки


```python
def fill_na_gaz(row):
    if row['Вздутие живота'] == 0:
        if row['Перистальтика'] == 4.0: return 3.0
        elif row['Перистальтика'] <=2.0: return  1.0
        elif row['Перистальтика'] ==3.0: return 2.0 
        else: return row['Вздутие живота']
    else: return row['Вздутие живота']
    
horses_no_null['Вздутие живота'] = horses_no_null.apply(fill_na_gaz, axis=1)
# horses_no_null.head(300)
horses_no_null[horses_no_null['Пульс'] == 0].head(150)
```

    2. на основе собственных выборок (см. "Пульс/Перистальтика")

В данном случае медианы на основе диапазонов по квантилям имеют больший разброс, а следовательно теоретически больше нам подходят (в теории, зависимость между пульсом и давлением, а значит и уровнем дегидратации/белка должна быть более явной).
Их будем использовать.

```python

def fill_na_protein_from_puls(row):
    """
    функция помогает найти значение в колонке 'Всего белка' на основании данных колонки 'Пульс'
    """
    if row['Всего белка'] == 0:
        if row['Пульс'] >= 30 and row['Пульс'] < 48: return 7.65
        elif row['Пульс'] >= 48 and row['Пульс'] < 64: return 7.5
        elif row['Пульс'] >= 64 and row['Пульс'] < 88: return 7.4
        elif row['Пульс'] >= 88: return 8.0
        else: return row['Пульс']
    else: return row['Всего белка']
    
horses_no_null['Всего белка'] = horses_no_null.apply(fill_na_protein_from_puls, axis=1)
horses_no_null[horses_no_null['Всего белка'] == 0.0].head(50)
# horses_no_null.head(300)
```

***Теперь разбираемся с Периферическим пульсом.***

Здесь все проще, поскольку зависимость между пульсом и периферическим пульсом гораздо выраженнее. Используем зависимость, обратную зависимости "Перферический пульс"/"Пульс"

У нас есть следующие данные:

Периферический пульс = 1:
Пульс, медиана: 52.0

Периферический пульс = 2:
Пульс, медиана: 60.0

Периферический пульс = 3:
Пульс, медиана: 88.0

Периферический пульс = 4:
Пульс, среднее: 91.4

в таком случае можно разбить на следующие диапазоны:
- если пульс в пределах до 56 (56 = 52 + ((60-52)/2)), то периферический пульс = 1
- если пульс в пределах от 57 до 74 (74 = 60 + ((88-60)/2), то периферический пульс = 2
- если пульс в пределах от 75 до 89,7 (89,7 = 75 + ((91,4-88)/2), то периферический пульс = 3
- если пульс в пределах от 89,8 и выше, то периферический пульс = 4

```python
def fill_na_pp_from_puls(row):
    """
    функция помогает найти значение в колонке 'Периферический пульс' на основании данных колонки 'Пульс'
    """
    if row['Периферический пульс'] == 0:
        if row['Пульс'] != 0:
            if row['Пульс'] <= 56: return 1.0
            elif row['Пульс'] >= 57 and row['Пульс'] <= 74: return 2.0
            elif row['Пульс'] >= 75 and row['Пульс'] < 89.7: return 3.0
            elif row['Пульс'] >= 89.8: return 4.0
            else: return row['Периферический пульс']
        else: return row['Периферический пульс']
    else: return row['Периферический пульс']
    
horses_no_null['Периферический пульс'] = horses_no_null.apply(fill_na_pp_from_puls, axis=1)
horses_no_null[horses_no_null['Периферический пульс'] == 0.0].head(50)
# horses_no_null.head(300)
```

Таким образом, на текущий момент у нас есть всего 5 строк, в которых по какой-либо колонке присутствуют значения = 0.
При этом:
- в 4 из 5 строк значения отсутствуют (равны 0) сразу в 4 колонках из 8
- в 1 строке значения отсутствуют в 2 колонках из 8 и оба этих значения исключительно категориальные, информирующие о наличии операции и итоге лечения.

Произвести корректную замену в подобном случае достаточно сложно, поскольку требуется искать более сложные и неявные взимосвязи между значениями столбцов (уверен, в теории это возможно, если ты ветеринар 85 lvl).

5 строк - это 1,67 % от всех данных, что достаточно мало для влияния на метрики, а следовательно этими строками можно пренебречь, просто удалив их.
