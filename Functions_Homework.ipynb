```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}


def input_person(): # ввод параметров 
    d_pers = input("Введите ФИО:\n ")
    return d_pers,


def input_number(): # ввод номера документа 
    d_num = input("Введите номер документа:\n")
    return d_num

def input_type(): # ввод  типа документа
    d_type = input("Введите тип документа:\n")
    return d_type

def input_shelve(): # ввод параметров 
    d_shelve = input("Введите номер полки:\n ")
    return d_shelve


def search_by_number(docs, d_num): 
    """функция, ищущая документ по номеру (документ есть - возвращается True, если нет - False)"""
    for el in docs:
        if d_num == el['number']:
            return True
    return False


def search_by_type(docs, d_type): 
    """функция, ищущая документ по типу (тип есть - возвращается True, если нет - False)"""
    for el in docs:
        if d_type == el['type']:
            return True
    return False

def search_by_shelve(dirs, d_shelve):
    for k, v in directories.items():
        if d_shelve in k:
            return True
    return False

def add_doc(dirs, docs, d_shelve, d_type, d_num, d_pers): 
    
    """функция, добавляющая документ на существующую полку.. Функция ничего не выводит на экран, 
    а только возвращает ряд значений. Такая реализация сделана для оптимизации решения задачи № 6, 7, 8""" 

    for k, v in dirs.items():
        if d_shelve in k:
            docs.append({'type': d_type, 'number': d_num, 'name': d_pers})
            dirs[d_shelve].append(d_num)
            

            
def delete_doc(dirs, docs, d_num,): 
    """функция, удаляющая документ на существующей полке. Функция ничего не выводит на экран, 
    а только возвращает ряд значений. Такая реализация сделана для оптимизации решения задачи № 8""" 
    for el in docs:
        for k, v in el.items():
            for s, n in dirs.items():
                if d_num in v and d_num in n:
                    docs.remove(el)
                    n.remove(d_num)

# Задание 1
def SearchByNumOut(docs): 
    """функция, выводящая результат работы search_bn(docs)"""
    d_num = input_number()
    #d_type = input_type()
    if search_by_number(docs, d_num): #and search_by_type(docs, d_type):
        for el in docs:
            if d_num == el['number']:
                print('Владелец документа: ', el['name'])
            
    else:
        print('Документ не найден в базе')
        

# Задание 2
def SearchInShelvesOut(dirs): # 
    """функция, выводящая номер полки по результатам работы функции search_in_shelves(dirs)"""
    finded, sf = search_in_shelves(dirs)
    if finded:
        print('Документ хранится на полке: ', sf)
    else:
        print('Документ не найден в базе')
        

# Задание 3
def data_output(dirs, docs):  
    """функция выводит итоговый список по документам владельцам и полкам"""
    for k, v in dirs.items():
        for el in docs:
            if el['number'] in v:
                el['shelve'] = k
                print('ФИО: ', el['name'], '. Тип документа:', el['type'], '. №:', el['number'], '. Номер полки:', el['shelve'], '.')                
        
        
        
# Задание 4
def dirs_quantity(dirs): 
    """ функция, выводящая количество полок"""
    d_q = ', '.join(list(dirs.keys()))
    return d_q

def add_shelve_command(dirs): 
    """ функция, добавляющая полку"""
    add_shelve = input("Введите номер полки:\n ")
    if add_shelve == None:
        return None, None
    if add_shelve not in dirs.keys():
        dirs[add_shelve] = []
        dirs_n = dirs_quantity(dirs)
        return True, dirs_n
    return False, None

def addshelve_out(dirs): 
    """Функция выводит результат по итогам добавления полки""" 
    finded, sf = add_shelve_command(dirs)
    if finded:
        print('Полка добавлена. Текущий перечень полок: ', sf)
    else:
        print('Такая полка уже есть, текущий перечень полок: ', dirs_quantity(dirs))        
    return dirs        

# Задание 5
    
def del_shelve_command(dirs):
    """Функция удаляет полку, если она пустая и если она существует""" 
    del_shelve = input("Введите номер полки:\n ")
    for k, v in dirs.items():
        if del_shelve in k:
            if not v:
                del(dirs[k])
                dirs_n = dirs_quantity(dirs)
                return True, dirs_n              
            else:
                dirs_n = dirs_quantity(dirs)
                return None, dirs_n
    return False, None


def delshelve_out(dirs):
    """Функция выводит результат по итогам удаления полки""" 

    finded, sf = del_shelve_command(dirs)
    if finded:
        print('Полка удалена. Текущий перечень полок: ', sf)
    elif sf:
         print('\033[91m\033[1mНа полке есть документы, удалите их перед удалением полки.\033[0m\033[0m')
    else:
        print('\033[91m\033[1mТакой полки не существует. Текущий перечень полок: \033[0m\033[0m', dirs_quantity(dirs))      
    return dirs
        

# Задание 6 Добавление документа       


def DocAddition(docs, dirs):
    """Функция добавляет документ на существующую полку и выводит итоговые сообщения в зависимости от результата""" 
    d_num =  input_number()
    d_type = input_type()
    if not search_by_number(docs, d_num) and not search_by_type(docs, d_type): 
        d_pers = input_person() 
        d_shelve = input_shelve()
        if not search_by_shelve(dirs, d_shelve):
            print('\033[91m\033[1m!!!Такой полки не существует!!!!\nСначала создайте полку командой "ads" \033[0m\033[0m')
        else:
            add_doc(dirs, docs, d_shelve, d_type, d_num, d_pers)
            print('Документ добавлен. Текущий список документов:\n')
            data_output(dirs, docs)
    else:
        for el in docs:
            if d_num == el['number'] and d_type == el['type']:
                print('\033[91m\033[1mCовпадение!!!\nВ базе уже есть человек с такими данными: \033[0m\033[0m', el['name'])
    return docs, dirs

                
                


# Задание 7 Удаление документа       
                                      
def DocTerminator(docs, dirs):
    """Функция, удаляет документ на существующую полку и выводит итоговые сообщения в зависимости от результата""" 
    d_num = input_number()
    if not search_by_number(docs, d_num): 
        print('\033[91m\033[1m!!!Такого документа нет!!!!\033[0m\033[0m')
    else:
        delete_doc(dirs, docs, d_num,)
        print('Документ удален. Текущий список документов:\n')
        data_output(dirs, docs)
    return docs, dirs

# Задание 8 Перемещение документа       

def docTransfer(docs, dirs):
    """Функция, сначала определяет наличие документа по номеру (с помощью search_by_number(docs, d_num),
    после чего, если такой документ есть удаляет его с текущего места и добавляет на указанную полку, если она существует.
    Если же полки нет, то выдается предупреждение с перечнем текущих полок""" 
    d_num = input_number()
    d_type = ''
    d_pers = ''
    for el in docs:
        if d_num == el['number']:
            d_type = el['type']
            d_pers = el['name']
    if not search_by_number(docs, d_num): 
        print('\033[91m\033[1m!!!Такого документа нет!!!!\nТекущий перечень документов\033[0m\033[0m')
        data_output(dirs, docs)
    else:
        d_shelve = input_shelve()
        if not search_by_shelve(dirs, d_shelve):
            print('\033[91m\033[1m!!!Такой полки нет!!!!\nТекущий перечень полок: \033[0m\033[0m', dirs_quantity(dirs))
        else:
            delete_doc(dirs, docs, d_num)
            add_doc(dirs, docs, d_shelve, d_type, d_num, d_pers)
            print('\033[91m\033[1mДокумент перемещен\nТекущий перечень документов\033[0m\033[0m')
            data_output(dirs, docs)
    return docs, dirs
    
def main():
    initial_input = input('введите команду: ')

    while initial_input != "x":
        if initial_input == "p":
            SearchByNumOut(documents)
            initial_input = input('введите команду: ')
        elif initial_input == "s":
            SearchInShelvesOut(directories)
            initial_input = input('введите команду: ')
        elif initial_input == "l":
            data_output(directories, documents)
            initial_input = input('введите команду: ')
        elif initial_input == "ads":
            addshelve_out(directories)
            initial_input = input('Введите команду: ')
        elif initial_input == "ds":
            delshelve_out(directories)
            initial_input = input('Введите команду: ')
        elif initial_input == "ad":
            DocAddition(documents, directories)
            initial_input = input('Введите команду: ')
        elif initial_input == "d":
            DocTerminator(documents, directories)
            initial_input = input('Введите команду: ')
        elif initial_input == "m":
            docTransfer(documents, directories)
            initial_input = input('Введите команду: ')
        else:
            print('Такой команды нет')
            initial_input = input('введите команду: ')
    if initial_input == "x":
            print('работа программы завершена')
main()
```
